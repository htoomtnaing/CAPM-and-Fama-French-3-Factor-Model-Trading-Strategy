```python
import pandas as pd
import numpy as np
import yfinance as yf
```

```python
# to create y3data.csv
# import corpfund csv, set index as ticker-calendar
cfdata = pd.read_csv('corpfund.csv', index_col = (0,2))
cfdata.head()
```

```python
# to create y3data.csv

# Keep rows with dimension ARQ
cfdata = cfdata[cfdata['dimension'] == 'ARQ']

# Create new truncated df using relevant columns: 'marketcap', 'pb'
tcf_df = cfdata[['marketcap', 'pb']]
tcf_df

# Top 1000 stocks by market cap
tickers = tcf_df.groupby('ticker').mean().sort_values('marketcap', ascending = False).iloc[0:2000]
tickers

# Filter df using relevant companies
tcf_df = tcf_df.loc[tickers.index]
tcf_df
```

```python
# to create y3data.csv

# Set start and end date
start_date = '2010-01-01'
end_date = '2020-12-31'

# Retrieve trading universe data from yahoo finance
ydata = pd.DataFrame(columns = ('Ticker', 'Date', 'Open','High','Low','Close','Adj Close','Volume'))
ydata = ydata.set_index(['Ticker', 'Date'])
for i in range(len(tickers.index)):
    ticker = tickers.index[i]
    data = yf.download(ticker, start_date, end_date)
    data = data.reset_index()
    data.insert(0, 'Ticker', ticker)
    data.set_index(['Ticker','Date'], inplace = True)
    ydata = pd.concat([ydata,data], axis = 0)
print(ydata)
```

```python
# to create y3data.csv

ydata = ydata.reset_index()
ydata
```

```python
# to create y3data.csv
ydata.to_csv('y3data.csv', index=False)
```

```python
# read y3data.csv 
full_yf = pd.read_csv('y3data.csv')
full_yf
```

```python
# tickers with 0 trading volumne on any day
zeroVolumeStocks = full_yf[full_yf['Volume'] == 0]['Ticker'].unique()

# extract rows
indexValues1 = full_yf[full_yf['Ticker'].isin(zeroVolumeStocks)].index

# drop rows with error stocks
full_yf = full_yf.drop(indexValues1)
full_yf
```

```python
len(full_yf['Ticker'].unique())
```

```python
# function to calculate daily returns based on tickers
def dailyReturns(df):
    int_col = ['Ticker', 'Close']
    close = df[int_col].values
    output = []
    for i in range(len(close)):
        if i == 0 or close[i][0] != close[i-1][0]:
            output.append(0.0)
        else:
            dailyReturn = (close[i][1] - close[i-1][1])/close[i-1][1]
            output.append(dailyReturn)
    df['Daily Returns'] = output
    
    return df

# Calculate Daily Returns
dailyReturns(full_yf)
full_yf.head()
```

```python
# tickers with Daily Returns > 1
highReturnStocks = full_yf[full_yf['Daily Returns'] > 1]['Ticker'].unique()

# extract rows
indexValues2 = full_yf[full_yf['Ticker'].isin(highReturnStocks)].index

# drop rows with error stocks
full_yf = full_yf.drop(indexValues2)
full_yf
```
